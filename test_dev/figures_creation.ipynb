```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/workspaces/baseline/exp/data_for_plot_creation/wandb_22_may_24.csv")
```

```python
df = df.rename(columns={'22/05/2024_16:35 - per_class/iou/total/background':"background", 
                   '22/05/2024_16:35 - per_class/iou/total/catenary_pole':"catenary_pole",
                   '22/05/2024_16:35 - per_class/iou/total/person':"person",
                   '22/05/2024_16:35 - per_class/iou/total/track':"track",
                   '22/05/2024_16:35 - per_class/iou/total/train':"train"})
```

```python
plot_data=df[["background","train","track","person","catenary_pole"]].loc[41]
```

```python
df[["background","train","track","person","catenary_pole"]].loc[0]
```

```python
fig, ax = plt.subplots()
bars= ax.bar(plot_data.index, plot_data.values)
ax.bar_label(bars, fmt="%.2f")
plt.title("Classwise Intersection over Union", fontdict={"size":14})
plt.xlabel("Class", fontdict={"size":14})
plt.ylabel("IOU", fontdict={"size":14})
plt.savefig("/workspaces/baseline/exp/figures/classwise_iou.jpeg", dpi=400)
```
