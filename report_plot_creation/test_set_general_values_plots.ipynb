```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/workspaces/baseline/exp/data_for_plot_creation/wandb_IoUs_TEST_SET_all_classes.csv")
```

```python
df
```

```python
df.columns = df.columns.str.replace("weight: online paste p=0.0 -", "baseline", regex=True)
df.columns = df.columns.str.replace("weight: online sparse p=0.9, r=40-50 -", "best_sparse",regex=True)
df.columns = df.columns.str.replace("weight: online paste p=0.8 -",  "best_paste",regex=True)
df.columns = df.columns.str.replace("per_class/iou/total/", "", regex=True)
df = df.drop(columns=[col for col in df.columns if 'MIN' in col or 'MAX' in col]) # Remove the useless columns that wandb generates
df.columns = df.columns.str.replace("/track", "m", regex=True)

```

```python
columns = [
    "background", "buffer_stop", "catenary_pole", "person", 
    "road_vehicle", "signal", "track", "train", "val/total/mIoU"
]
run_values = np.vstack((df.filter(regex="baseline").iloc[0,1:].values*100,
                       df.filter(regex="best_sparse").iloc[0,1:].values*100, 
                       df.filter(regex="best_paste").iloc[0,1:].values*100))
# Create the DataFrame
values_df = pd.DataFrame(run_values, columns=columns)
```

```python
desired_order = ["background", "person", "train", "road_vehicle", "track", "catenary_pole", "signal", "buffer_stop", "val/total/mIoU"]
values_df=values_df[desired_order]
values_df.round(2)
```
